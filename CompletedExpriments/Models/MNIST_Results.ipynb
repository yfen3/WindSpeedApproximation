```python
teacher_test_accuracy = 0.9802
```

```python
test_accuracy_with_distillation = [
    0.9677000045776367,
    0.9009000062942505,
    0.9799000024795532,
    0.972599983215332,
    0.9555000066757202
]
```

```python
number_of_experiments = len(test_accuracy_with_distillation)
print(number_of_experiments)
```

```python
x = range(0, number_of_experiments)

plt.figure(figsize=(8,8))
plt.title('Test Accuracy with and without distillation') 

ax = plt.gca()
# ax.set_xlim([0, 1])
# ax.set_ylim([0, 1])


plt.plot(x, test_accuracy_with_distillation,'d-')
plt.plot(x, test_accuracy_without_distillation,'x-')
plt.axhline(y=teacher_test_accuracy, color='r', linestyle='--')

plt.legend(['With Distillation', 'Without Distillation', 'Teacher Model Baseline'])

plt.xlabel('Experiment')
plt.ylabel('Accuracy')
```
