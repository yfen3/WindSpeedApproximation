```python
# Get the data for expeirment
import sklearn
from sklearn.datasets import fetch_california_housing
import pandas as pd 
import glob
import os
```

```python
#source: https://www.geeksforgeeks.org/how-to-read-all-csv-files-in-a-folder-in-pandas/
#source: https://stackoverflow.com/questions/27749553/how-can-i-read-the-contents-of-all-the-files-in-a-directory-with-pandas
path = os.getcwd()
csv_files = glob.glob(os.path.join(path+'\Data\AB_wind_data', "*.csv"))
```

```python
l = [pd.read_csv(file, encoding="ISO-8859-1") for file in csv_files]
df = pd.concat(l, axis=0)
```

```python
print(len(df['Station Name'].unique()))
print(df['Station Name'].unique())
```

```python
df
```

```python
df.isnull().sum()
```

```python
col_to_use = {
    'Longitude (x)' : 'longitude',
    'Latitude (y)': 'latitude',
    'Station Name' : 'name',
    'Climate ID':'id',
    'Date/Time (LST)':'date',
    'Temp (°C)':'temp',
    'Precip. Amount (mm)':'precip',
    'Wind Dir (10s deg)':'wind_direction',
    'Wind Spd (km/h)':'wind_speed'
}
```

```python
null_mask = df['Wind Spd (km/h)'].isnull()
null_rows = df[null_mask]

print(null_rows)
```

```python
dropped_df = df.dropna(subset=['Wind Spd (km/h)'])
```

```python
dropped_df
```

```python
processed_df = dropped_df[col_to_use.keys()].copy()
```

```python
processed_df = processed_df.rename(columns=col_to_use)
```

```python
processed_df.to_csv('Data\processed_ab_wind.txt', sep = ',', index = False)
```
